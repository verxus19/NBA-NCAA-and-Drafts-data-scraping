```python
class_converter_dic = {'Fr *': 1, 'Fr': 1, 
                       'So *': 2, 'So': 2, 
                       'Jr *': 3, 'Jr': 3,
                       'Sr *': 4, 'Sr': 4}
class_converter = lambda x: class_converter_dic[x] if x in class_converter_dic.keys() else 0
```

```python
plt.scatter(draft_2007.Pick, draft_2007.MPG)
plt.show()
```

```python
plt.scatter(draft_2007.Pick, draft_2007.Class.map(class_converter))
plt.show()
```
